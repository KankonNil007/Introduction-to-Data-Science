```python
import numpy as np
import math

```

```python
a = np.array([10, 20, 30, 40])
b = np.array([1, 2, 3, 4])

c = a + b
print(c)

c = a - b
print(c)

c = a * b
print(c)

c = a / b
print(c)
```

```python
farenheit = np.array([5, 0, -4, 23, -10])

celcius = (farenheit - 32) * 5/9

print(celcius)
```

```python
print(celcius > -20)
print(celcius%2 == 0)
```

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[3, 2], [0, 4]])

print(A*B) # Element-wise Multiplication
print(A@B) # Dot Product
```

```python
arr1 = np.array([1, 2, 3, 4])
print(arr1.dtype)

arr2 = np.array([10.4, 2.7, 9.3, 6.4])
print(arr2.dtype)
```

```python
arr3 = arr1 + arr2
print(arr3)
print(arr3.dtype)
```

```python
print(arr3.sum())
print(arr3.max())
print(arr3.min())
print(arr3.mean())
```
