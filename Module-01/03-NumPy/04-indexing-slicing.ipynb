```python
a = np.array([3,4,2,45,23,65,87])
print(a[2:4])
print(a[2:])
print(a[:4])
```

```python
a = np.array([[1,2,34,4], [4,5,6,7], [6,7,8,9]])
print(a)
```

```python
print(a[:2]) # rows
print(a[:2, 1:3]) # rows, columns
```

```python
sub_array = a[:2, 1:3]
print("sub array index [0,0] value before change:", sub_array[0,0])
sub_array[0,0] = 50
print("sub array index [0,0] value after change:", sub_array[0,0])
print("original array index [0,1] value after change:", a[0,1])
```
