```python
import numpy as np
from PIL import Image
from IPython.display import display
```

```python
im = Image.open("images/grayscaled.png")
display(im)
```

```python
imArr = np.array(im)
print(imArr.shape)
print(imArr)
```

```python
mask =  np.full(imArr.shape, 255)
modArr = imArr - mask
modArr = modArr * -1

modArr = modArr.astype(np.uint8)
print(modArr)
```

```python
display(Image.fromarray(modArr))
```

```python
reshaped = np.reshape(modArr, (400, 625))
print(reshaped.shape)
display(Image.fromarray(reshaped))
```
