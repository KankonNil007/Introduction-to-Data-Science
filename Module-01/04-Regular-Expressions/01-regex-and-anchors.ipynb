```python
import re
```

```python
text = "This is a good day"

if re.search("good", text):
    print("Wonderful")
else:
    print("Alas!!")
```

```python
text = "Amy works diligently. Amy gets good grades. Our student Amy is successful."

print(re.split("Amy", text))

print(re.findall("Amy", text))
```
