```python
import re
```

```python
grades = "ABAAAABBCBABACBA"

print(re.findall("A{2,10}", grades))
```

```python
print(re.findall("A{1,1}A{1,1}", grades))
```

```python
print(re.findall("AA", grades))
```

```python
print(re.findall("A{2}", grades))
```

```python
print(re.findall("A{1,10}B{1,10}C{1,10}", grades))
```

```python
# an asterix * to match 0 or more times, a question mark ? to match one or more times, or a + plus sign to match one or more times.
```

```python
with open("datasets/ferpa.txt", "r") as file:
    text = file.read()

print(text)
```

```python
print(re.findall(r"[a-zA-Z]{1,100}\[edit\]", text))
```

```python
print(re.findall(r"[\w]{1,100}\[edit\]", text)) 
# \w means any characters including digits
```

```python
print(re.findall(r"[\w]*\[edit\]", text)) 
# * means 0 or more times
```

```python
print(re.findall(r"[\w ]*\[edit\]", text)) 
```

```python
for title in re.findall(r"[\w ]*\[edit\]", text):
    print(re.split(r"[\[]",title)[0])
```
