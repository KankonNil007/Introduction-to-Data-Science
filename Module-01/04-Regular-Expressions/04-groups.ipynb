```python
import re
```

```python
with open("datasets/ferpa.txt", "r") as file:
    text = file.read()

print(text)
```

```python
re.findall(r"([\w ]*)(\[edit\])", text)
```

```python
for item in re.finditer(r"([\w ]*)(\[edit\])", text):
    print(item.groups())
```

```python
for item in re.finditer(r"([\w ]*)(\[edit\])", text):
    print(item.group(1))
```

```python
for item in re.finditer(r"(?P<title>[\w ]*)(?P<edit_link>\[edit\])", text):
    print(item.groupdict()['title'])
```

```python
print(item.groupdict())
```
