```python
import re
```

```python
grades = "ABAAAABBCBABACBA"

print(re.findall("B", grades))
print(re.findall("[AB]", grades))
```

```python
print(re.findall("[A-C]", grades))
print(re.findall("[A][B-C]", grades))
```

```python
print(re.findall("AB|AC", grades))
```

```python
print(re.findall("[^A]", grades)) # skips all As
```

```python
re.findall("^[^A]",grades)
```
