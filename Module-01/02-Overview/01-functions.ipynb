```python
def add_numbers(a, b, c=None):
    if (c == None):
        return a + b
    else:
        return a + b + c

print(add_numbers(1, 2, 3))
```

```python
def do_math(a, b, c="add"):
    if (c == "add"):
        return a + b
    else:
        return a - b

print(do_math(1, 2))
```

```python
def add_numbers(x, y, z=None, flag=False):
    if (flag):
        print('Flag is true!')
    if (z == None):
        return x + y
    else:
        return x + y + z


print(add_numbers(1, 2, 4, flag=True))
```
