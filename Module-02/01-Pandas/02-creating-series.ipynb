```python
import pandas as pd
```

```python
subjects = {
    "Alice": "Physics",
    "John": "Chemistry",
    "Molly": "Mathematics"
}

ds = pd.Series(subjects)

print(ds)
```

```python
print(ds.index)
```

```python
students = [("Alice","Brown"), ("Jack", "White"), ("Molly", "Green")]
ds = pd.Series(students)

print(ds)
```

```python
ds = pd.Series(["Alice", "Molly", "Jack"], index=["Physics", "Chemistry", "Mathematics"])

print(ds)
```

```python
subjects = {
    "Alice": "Physics",
    "John": "Chemistry",
    "Molly": "Mathematics"
}

ds = pd.Series(subjects, index=["Alice", "Molly", "Sam"])

print(ds)
```
