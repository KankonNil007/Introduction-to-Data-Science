```python
import pandas as pd
```

```python
students_classes = {'Alice': 'Physics',
                   'Jack': 'Chemistry',
                   'Molly': 'English',
                   'Sam': 'History'}
ds = pd.Series(students_classes)
print(ds)
```

```python
# Indexing by Number or Entry
print(ds.iloc[3])

# Indexing by index names
print(ds.loc["Molly"])
```

```python
# Indexing Shortcuts - (recomended not to use)
print(ds[3])
print(ds["Molly"])
```

```python
class_code = {99: 'Physics',
              100: 'Chemistry',
              101: 'English',
              102: 'History'}
ds = pd.Series(class_code)

print(ds)
```

```python
print(ds.iloc[2])
print(ds.loc[99])
```
