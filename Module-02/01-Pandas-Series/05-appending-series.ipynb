```python
import pandas as pd
```

```python
ds = pd.Series([1, 2, 3])

ds.loc["History"] = 104

print(ds)
```

```python
students_classes = pd.Series({'Alice': 'Physics',
                   'Jack': 'Chemistry',
                   'Molly': 'English',
                   'Sam': 'History'})
print(students_classes)
```

```python
kelly_classes = pd.Series(['Philosophy', 'Arts', 'Math'], index=['Kelly', 'Kelly', 'Kelly'])
print(kelly_classes)
```

```python
allStudClasses = pd.concat([students_classes, kelly_classes])

print(allStudClasses)
```
